```python
#4.Proponer 5 ejercicios de la tabla Player
query = """SELECT COUNT (DISTINCT(Player_name))AS TotalJugadores
FROM Player; """
Total = pd.read_sql(query,conn)
Total
```

```python
#Promeido de altura de los jugadores
query = """SELECT AVG (DISTINCT(height))AS Promedio_altura
FROM Player"""#AS seaplica como alias para darle un orden y titulo a la tabla
Total = pd.read_sql(query,conn)
Total
```

```python
#Juagador con peso Max
query = """SELECT MAX (DISTINCT(weight))AS Peso_max, Player_name AS Player_name
FROM Player"""#AS seaplica como alias para darle un orden y titulo a la tabla
Total = pd.read_sql(query,conn)
Total
```

```python
#Promeido de peso de los jugadores
query = """SELECT AVG (DISTINCT(WEIGHT))AS Promedio_PESO
FROM Player"""#AS seaplica como alias para darle un orden y titulo a la tabla
Total = pd.read_sql(query,conn)
Total
```

```python
4.1.2#Traer el numero de etapas jugadas de la tabla match
query = """SELECT COUNT (DISTINCT STAGE) FROM Match;"""
stages = pd.read_sql(query,conn)
stages
#en todas las ligas europeas entre el ao 2008 a 2016 se jugaron 38 temporadas
```

```python
#4.1.3 Prctica: Traer la suma de goles hechos por los equipos visitantes de la tabla Match
query = """SELECT SUM (DISTINCT (away_team_goal)) AS suma_de_goles FROM Match """
suma_goles = pd.read_sql(query,conn)
suma_goles
```

```python
#4.1.4 Prctica: Traer el promedio de goles hechos por los equipos locales y
#el promedio de goles hechos por los visitantes de la tabla Match
query = """SELECT AVG (home_team_goal),AVG (away_team_goal)
FROM Match"""
#INNER JOIN (away_team_goal)
#ON Match.home_team_goal=Match.away_team_goal;
prom = pd.read_sql(query,conn)
prom
```

```python
#4.1.5 Prctica: Traer la mxima cantidad de goles que un equipo visitante anot en los juegos de la tabla Match
query = """SELECT MAX(away_team_goal) FROM Match"""
Max_goles = pd.read_sql(query,conn)
Max_goles
```

```python
#Traer el numero de etapas jugadas de la tabla match del ao 2015/2016
query = """SELECT * FROM Match WHERE NOT SEASON = '2015/2016'; """# TRAE SOLO LAS TEMPORADAS 2015/2016
stages1 = pd.read_sql(query,conn)
stages1
```

```python

# se desea conocer el pais al que corresponde cada liga
query = """SELECT L.id AS Id_league,
           L.name AS Nombre_League,
           C.name AS Pais
FROM League AS L
JOIN Country AS C ON C.id = L.country_id;"""#selecciona varias columnas dando un nuevo nombre y luego las une todas en una tabla

Countrys_leagues = pd.read_sql(query, conn)
Countrys_leagues
```

```python
# otra opcion...
query = """SELECT * FROM League JOIN Country
           ON Country.id = League.country_id; """

country_leagues = pd.read_sql(query,conn)
country_leagues
```
