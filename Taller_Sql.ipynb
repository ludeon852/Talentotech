```python
#Proponer 5 ejercicios de la tabla Player
query = """SELECT COUNT (DISTINCT(Player_name))AS TotalJugadores
FROM Player; """
Total = pd.read_sql(query,conn)
Total
```

```python
#Promeido de altura de los jugadores
query = """SELECT AVG (DISTINCT(height))AS Promedio_altura
FROM Player"""#AS seaplica como alias para darle un orden y titulo a la tabla
Total = pd.read_sql(query,conn)
Total
```

```python
#Juagador con peso Max
query = """SELECT MAX (DISTINCT(weight))AS Peso_max, Player_name AS Player_name
FROM Player"""#AS seaplica como alias para darle un orden y titulo a la tabla
Total = pd.read_sql(query,conn)
Total
```

```python
#Promeido de peso de los jugadores
query = """SELECT AVG (DISTINCT(WEIGHT))AS Promedio_PESO
FROM Player"""#AS seaplica como alias para darle un orden y titulo a la tabla
Total = pd.read_sql(query,conn)
Total
```

```python
#Traer el numero de etapas jugadas de la tabla match
query = """SELECT COUNT (DISTINCT STAGE) FROM Match;"""
stages = pd.read_sql(query,conn)
stages
#en todas las ligas europeas entre el ao 2008 a 2016 se jugaron 38 temporadas
```

```python
#Traer el numero de etapas jugadas de la tabla match del ao 2015/2016
query = """SELECT * FROM Match WHERE NOT SEASON = '2015/2016'; """# TRAE SOLO LAS TEMPORADAS 2015/2016
stages1 = pd.read_sql(query,conn)
stages1
```

```python

# se desea conocer el pais al que corresponde cada liga
query = """SELECT L.id AS Id_league,
           L.name AS Nombre_League,
           C.name AS Pais
FROM League AS L
JOIN Country AS C ON C.id = L.country_id;"""#selecciona varias columnas dando un nuevo nombre y luego las une todas en una tabla

Countrys_leagues = pd.read_sql(query, conn)
Countrys_leagues
```
